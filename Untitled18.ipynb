```python
##the model we will using is logistic regression
##step 3 is feature selection
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)#10% test and 90%training
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
##scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(X_train,y_train)#model trainig
```

```python
##evaluate the model
y_pred=clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
##Now to see how it has split the datasets with line so
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.values,clf=clf,legend=2)
```
